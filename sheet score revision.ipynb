### Removing white spaces from values and converting to date-time

```python
subject_id_list = [i.strip() for i in list(current_sheet['Subject ID\xa0'])]
subject_scan_date = [i.strip()  for i in list(current_sheet['Study Date\xa0'])]
current_sheet['Subject ID\xa0'] = subject_id_list
current_sheet['Study Date\xa0'] = subject_scan_date

current_sheet['Study Date\xa0'] =  pd.to_datetime(current_sheet['Study Date\xa0'], format='%m/%d/%Y')
adnimerge_req_df['EXAMDATE'] = pd.to_datetime(adnimerge_req_df['EXAMDATE'], format='%Y-%m-%d')
```

```python
subject_id_list = list(current_sheet['Subject ID\xa0'])
subject_scan_date = list(current_sheet['Study Date\xa0'])
print(subject_id_list[1],subject_scan_date[1])
```

```python
def closest_date(scan_date,list_dates):
    diff=timedelta(100000)#approx 273 years
    index_closest_date = 0
    for i in range(len(list_dates)):
        if abs(scan_date-list_dates[i])<diff:
            diff = abs(scan_date-list_dates[i])
            index_closest_date = i
    return list_dates[index_closest_date]
```

## Trail run for 10 rows

```python
revised_scores = []

assert (len(subject_id_list)==len(subject_scan_date))

for i in range(len(subject_scan_date[:10])):
    exam_df = adnimerge_req_df[(adnimerge_req_df['PTID'] == subject_id_list[i])&
                               (adnimerge_req_df['EXAMDATE']>subject_scan_date[i]-timedelta(days=150))&
                               (adnimerge_req_df['EXAMDATE']<subject_scan_date[i]+timedelta(days=150))]
    print(subject_scan_date[i])
    print(exam_df)
    print(exam_df.shape)
    if exam_df.shape[0]>1:
        print("*******************************")
        exam_dates_in_range = list(exam_df['EXAMDATE'])
        nearest_date = closest_date(subject_scan_date[i],exam_dates_in_range)
        print("closest date = ",nearest_date)
        final_exam_df = exam_df[exam_df['EXAMDATE']==nearest_date]
        print("*******************************")
        print(final_exam_df)
        print("*******************************")
        assert (final_exam_df.shape[0]==1)
        for index, rows in final_exam_df.iterrows():
            revised_scores.append([rows.EXAMDATE, rows.APOE4, rows.ADAS11, rows.ADAS13, rows.MMSE, rows.CDRSB,
                                   rows.RAVLT_immediate, rows.RAVLT_learning,rows.RAVLT_forgetting,
                                   rows.LDELTOTAL, rows.DIGITSCOR, rows.EcogPtTotal])
    elif exam_df.shape[0]==1:
        for index, rows in exam_df.iterrows():
            revised_scores.append([rows.EXAMDATE, rows.APOE4, rows.ADAS11, rows.ADAS13, rows.MMSE, rows.CDRSB,
                                   rows.RAVLT_immediate, rows.RAVLT_learning,rows.RAVLT_forgetting,
                                   rows.LDELTOTAL, rows.DIGITSCOR, rows.EcogPtTotal])
    else:
        revised_scores.append(['Not found']*12)

print(len(revised_scores))
print(revised_scores)
```

## Running on all rows

```python
revised_scores = []

assert (len(subject_id_list)==len(subject_scan_date))

for i in range(len(subject_scan_date)):
    exam_df = adnimerge_req_df[(adnimerge_req_df['PTID'] == subject_id_list[i])&
                               (adnimerge_req_df['EXAMDATE']>subject_scan_date[i]-timedelta(days=150))&
                               (adnimerge_req_df['EXAMDATE']<subject_scan_date[i]+timedelta(days=150))]
    
    if exam_df.shape[0]>1:
        exam_dates_in_range = list(exam_df['EXAMDATE'])
        nearest_date = closest_date(subject_scan_date[i],exam_dates_in_range)
        final_exam_df = exam_df[exam_df['EXAMDATE']==nearest_date]
        
        assert (final_exam_df.shape[0]==1)
        for index, rows in final_exam_df.iterrows():
            revised_scores.append([rows.EXAMDATE, rows.APOE4, rows.ADAS11, rows.ADAS13, rows.MMSE, rows.CDRSB,
                                   rows.RAVLT_immediate, rows.RAVLT_learning,rows.RAVLT_forgetting,
                                   rows.LDELTOTAL, rows.DIGITSCOR, rows.EcogPtTotal])
    elif exam_df.shape[0]==1:
        for index, rows in exam_df.iterrows():
            revised_scores.append([rows.EXAMDATE, rows.APOE4, rows.ADAS11, rows.ADAS13, rows.MMSE, rows.CDRSB,
                                   rows.RAVLT_immediate, rows.RAVLT_learning,rows.RAVLT_forgetting,
                                   rows.LDELTOTAL, rows.DIGITSCOR, rows.EcogPtTotal])
    else:
        revised_scores.append(['Not found']*12)
```

```python
current_sheet[['Behavioural_EXAMDATE', 'APOE4_revised', 'ADAS11_revised', 'ADAS13_revised', 'MMSE_revised', 'CDRSB_revised',
               'RAVLT_immediate_revised', 'RAVLT_learning_revised', 'RAVLT_forgetting_revised',
               'LDELTOTAL_revised', 'DIGITSCOR_revised', 'EcogPtTotal_revised']] = pd.DataFrame(revised_scores,index=current_sheet.index)
```
